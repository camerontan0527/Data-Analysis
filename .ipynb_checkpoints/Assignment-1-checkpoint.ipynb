## Introduction

**Bicycle Sharing Systems** are a healthy, cheap and green way to navigate between one place to another. Now with the new generation of electronic registration and sharing, the whole process of transportation, right from the membership to the rental and returning back has become convenient and automatic. Through these systems, user is able to easily rent a bike from a particular position and return back at another position. Bike sharing companies like **Mobike**, **OBike** and **AirBike** have grown pretty popular throughout Australia in the past few years, due to them being environment-friendly, pro-health and not to mention, a cheap and convenient mode of transport.

Apart from interesting real world applications of bike sharing systems, the characteristics of data being generated by these systems make them attractive for research. Opposed to other transport services such as bus or subway, the **duration of travel**, **departure** and **arrival position** is explicitly recorded in these systems. This feature turns bike sharing system into a virtual sensor network that can be used for sensing mobility in the city. Hence, it is expected that most of important events in the city could be detected via monitoring these data.

### 4. What relationship can you find among the attributes relating to weather in this dataset?
<span style= 'float: right;'><b>[6 marks]</b></span>

```python
# YOUR CODE HERE

temp_list = df_bike.temp.values
normal_temp = temp_list * 41
atemp_list = df_bike.atemp.values
normal_atemp = atemp_list * 50
dif = normal_temp - normal_atemp
dif = [abs(elem) for elem in dif]
df_temp = df_bike.copy()
df_temp['temp_dif'] = dif
df_temp_dif = df_temp.groupby('weathersit')['temp_dif'].mean()
clear = df_temp[df_temp.weathersit == 1]
mist = df_temp[df_temp.weathersit == 2]
rain = df_temp[df_temp.weathersit == 3]
fig2 = plt.figure(figsize = (24, 16))
ax5 = fig2.add_subplot(221)
ax5 = df_temp_dif.plot(kind = 'bar')
ax5.set_title('Average gap between temperature and feeling temperature in different weather situations', fontsize = 22)
ax5.set_ylabel('Temperature gap in Celsius', fontsize = 15)
ax6 = fig2.add_subplot(222)
ax6.boxplot([list(clear.temp_dif.values), list(mist.temp_dif.values), list(rain.temp_dif.values)], labels = ['Clear', 'Mist', 'Light Rain/Snow'])
ax6.set_title('Temperature gap in different weather situations', fontsize = 22)
ax6.set_ylabel('Temperature gap in Celsius', fontsize = 15)
ax6.legend()

# (ADD ANY ADDITIONAL CELLS AS REQUIRED)
```

<span style="color:blue">
    
### YOUR ANSWER HERE

According to the two plots above, we can find out the realtionship between the temperature gap and different weather situations. The two plots indicate that the largest gap between actual temperature and feeling temperature was in clear days. Then comes that in mist days. The smallest gap occured in light rain/snow days. 


</span>

## LINEAR REGRESSION

Now comes the fun part! In this section you have to implement a **Linear Regression Model**. Companies like **Mobike** and **AirBike** would like to make some predictions on their sales for the next few years. A linear Regression Model will help them predict the growth of their company in the coming years.

Your job is to implement a Linear Regression Model that predicts the number of bikes rented daily, based on the environmental and seasonal variables. There are multiple ways of doing this in Python. You are free to use any Python modules you like for this implementation. 
<span style= 'float: right;'><b>[25 marks]</b></span>

```python
# YOUR CODE HERE

lm = LinearRegression()
x = df_bike[['season', 'weathersit', 'temp', 'atemp', 'hum', 'windspeed']]
y = df_bike.cnt
lm.fit(x, y)
coef = lm.coef_
intercept = lm.intercept_
print('The coefficients of the model are:\n')
print(coef)
print('\nThe intercept of the model is:\n')
print(intercept)
print('\nSo the formula of this model is:\n')
print(str(intercept) + ' + (' + str(coef[0]) + ') * season + (' + str(coef[1]) + ') * weathersit + (' + str(coef[2]) + ') * temp + (' + str(coef[3]) + ') * atemp + (' + str(coef[4]) + ') * hum + (' + str(coef[5]) + ') * windspeed') 
season = np.array(df_bike['season']).reshape([-1, 1])
weathersit = np.array(df_bike['weathersit']).reshape([-1, 1])
temp = np.array(df_bike['temp']).reshape([-1, 1])
atemp = np.array(df_bike['atemp']).reshape([-1, 1])
hum = np.array(df_bike['hum']).reshape([-1, 1])
windspeed = np.array(df_bike['windspeed']).reshape([-1, 1])
estimate = intercept + coef[0] * season + coef[1] * weathersit + coef[2] * temp + coef[3] * atemp + coef[4] * hum + coef[5] * windspeed
print("\nThe accuracy(R^2) of this model is:\n")
print(skl.metrics.r2_score(df_bike['cnt'], estimate))

# (ADD ANY ADDITIONAL CELLS AS REQUIRED)
```
